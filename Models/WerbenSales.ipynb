```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import catboost as cb
```

```python
df = pd.read_csv('Walmart.csv')
df.drop(columns=['Date', 'Store'])
#print(df.head()) 'Weekly_Sales'
X = df.loc[:, ['Temperature','Fuel_Price','CPI','Unemployment']]
y = df.loc[:, ['Weekly_Sales']]
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#from xgboost import XGBRegressor
#regressor = XGBRegressor()
#regressor.fit(X_train, y_train)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.values.reshape(len(y_test),1)),1))
```

```python
train_dataset = cb.Pool(X_train, y_train) 
test_dataset = cb.Pool(X_test, y_test)
```

```python
model = cb.CatBoostRegressor(loss_function='RMSE')
```

```python
grid = {'iterations': [100, 150, 200],
        'learning_rate': [0.03, 0.1],
        'depth': [2, 4, 6, 8],
        'l2_leaf_reg': [0.2, 0.5, 1, 3]
        }
model.grid_search(grid, train_dataset)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.inspection import permutation_importance

pred = model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(y_test, pred)))
r2 = r2_score(y_test, pred)
LinRegr2 = r2_score(y_test,y_pred)
print("Testing performance for catboost")
print('RMSE: {:.2f}'.format(rmse))
print('R2: {:.2f}'.format(r2))
print('Linear Regression R2: {:.2f}'.format(LinRegr2))
```
